The degree centrality is defined as the number of connections a node has to other nodes,ie it shows the connectedness of the node.

The eigenvector centrality is the amount of influence a node has in a network,if it is connected to other nodes which are also of high influence then it has a higher score.

The difference in them is that the degree centrality just assesses the amount of connections a node has, wether they be of high value or not, the eigenvector centrality also assesses the quality of the connections of the node in the network.

A use case for the degree centrality could be for selecting salesperson or human resource managers as they would have more number of connections and more connections have a higher probability of landing a customer or a potential candidate.
For the eigenvector centrality a use case could be to select a media anchor or related person to convey an important message to the general public as that person would have better influence to other media persons and hence spreading the message would be easier.

```
from matplotlib.pyplot import figure
def plotshortest(source,target):
  figure(num=None, figsize=(18, 18), dpi=80)
  p = nx.shortest_path(Gx,source,target)
  
  for edge in Gx.edges():
    Gx[edge[0]][edge[1]]['color'] = 'black'
    
  for i in range(len(p)-1):
    Gx[p[i]][p[i+1]]['color'] = 'red'
    
  edge_color_list = [ Gx[e[0]][e[1]]['color'] for e in Gx.edges() ]
  nx.draw(Gx,edge_color = edge_color_list, with_labels = True)
  plt.show()
```


The Girvan newman algorithm uses edge betweenness, which is calculated by the number or shortest paths between pairs of nodes along the edge for which the betweennes is being calculated.

The edges which connect communities will have high betweenness and hence the algorithm removes these edge progressively and the connected components of the graph are communities.

The algorithm is as follows:

1.   calculate betweenness of all edges
2.   remove edge with highest betweenness


repeat until no edges remain

```

from networkx.algorithms import community

communities = community.girvan_newman(Gx)
topLevelCommunities = next(communities)
nextlevelCommunities = next(communities)
toplot = sorted(map(sorted, nextlevelCommunities))
```

```
len(toplot)
```

```
color_map = []
figure(num=None, figsize=(18, 18), dpi=80)
for node in Gx:
  if node in toplot[0]:
    color_map.append('blue')
  elif node in toplot[1]:
    color_map.append('red')
  elif node in toplot[2]:
    color_map.append('black')
  else:
    color_map.append('green')



nx.draw(Gx, node_color=color_map, with_labels=True)
plt.show()
```
